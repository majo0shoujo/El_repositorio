# Importando librerias 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Leer imagen 

```python
img = cv2.imread('Img_1.jpg', 0)
img2 = img.copy()
```

# Dimension de la Imagen

```python
(row, col) = img.shape
```

# Valor mínimo y máximo para la intensidad de píxeles

```python
for i in range(row):
    for j in range(col):
        if (img[i][j] < 10):
            img2[i][j] = 10
        if (img[i][j] > 240):
            img2[i][j] = 240
```

# Normalizar imagen

```python
f_max = img2.max()
f_min = img2.min()
img3 = img2.copy()
for i in range(row):
    for j in range(col):
        img3[i][j] = ((img2[i][j] - f_min) / (f_max - f_min)) * 256
```

 # Celdas imagen_ histograma

```python
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(8, 11)) 

axs[0][0].imshow(img, cmap="gray")
axs[0][0].axis("off")

axs[0][1].hist(img.ravel(), 256, [0, 256], color='gray')
axs[0][1].set_title('Histograma imagen 1')
axs[0][1].axis("off")


axs[1][0].imshow(img2, cmap="gray")
axs[1][0].axis("off")

axs[1][1].hist(img2.ravel(), 256, [0, 256], color='gray')
axs[1][1].set_title('Histograma imagen 2')
axs[1][1].axis("off")


axs[2][0].imshow(img3, cmap="gray")
axs[2][0].axis("off")

axs[2][1].hist(img3.ravel(), 256, [0, 256], color='gray')
axs[2][1].set_title('Histograma imagen 3')
axs[2][1].axis("off")
```
