# Librerias <3



```python
import cv2
import matplotlib.pyplot as plt
```

# Cargar imagen


```python
img = cv2.imread('img_1.jpeg')
```

# La imagen reducida

```python
img_reducida = cv2.resize(img, None, fx=0.5, fy=0.5)
```


##  La imagen a blanco y negro

```python
img_byw = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

```

# La imagen invertida

```python
img_invertida = cv2.flip(img, 1)
```

# La imagen con Detección de bordes a la imagen reducida

```python
img_bordes = cv2.Canny(img_reducida, 100, 200)
```

# Mostrar las imágenes

```python
fig, axs = plt.subplots(4, 2, figsize=(10,10))

axs[0, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title('Imagen original')

axs[0, 1].imshow(cv2.cvtColor(img_reducida, cv2.COLOR_BGR2RGB))
axs[0, 1].set_title('Imagen reducida')

axs[1, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[1, 0].set_title('Imagen original')

axs[1, 1].imshow(img_byw, cmap='gray')
axs[1, 1].set_title('Imagen en blanco y negro')

axs[2, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[2, 0].set_title('Imagen original')

axs[2, 1].imshow(cv2.cvtColor(img_invertida, cv2.COLOR_BGR2RGB))
axs[2, 1].set_title('Imagen invertida')


axs[3, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[3, 0].set_title('Imagen original')

axs[3, 1].imshow(cv2.cvtColor(img_bordes, cv2.COLOR_BGR2RGB))
axs[3, 1].set_title('Imagen reducida con filtro de detección de bordes')
```
